```python
import sklearn.datasets
import pandas as pd
import numpy as np
import umap
import matplotlib as plt
```

```python
data = pd.read_csv("/home/ona/Downloads/Project/send/Daegu_RE.csv")
```

```python
numeric = data._get_numeric_data()
```

```python
import umap.plot
```

```python
mapper = umap.UMAP(n_neighbors=30,
                      min_dist=0.1).fit(numeric)
```

```python
umap.plot.points(mapper, labels=data.iloc[:,16], color_key_cmap='cet_CET_R4')
```

```python
data.iloc[:,16]
```

```python
hover_data = pd.DataFrame({
                           'price':data.iloc[:,0],
                           'apt-management-type':data.iloc[:,8],
                           'subway-station':data.iloc[:,16],
                          })
```

```python
umap.plot.output_notebook()
```

```python
p = umap.plot.interactive(mapper, labels=data.iloc[:,16][:30000], hover_data=hover_data, point_size=2, color_key_cmap='cet_CET_R4')
umap.plot.show(p)
```
